 # Q2)  For numerical stability i took cfl number to be 0.2 whiach gives us the dt as 0.0002 
1. Note: run first three sections before running for any scheme 

```python
#input paramenters
n = 1001
x0 = 0
x1 = 1
dx =(x1-x0)/(n-1)
cfl = 0.2
c=1
u = np.zeros(n)
u_old = np.zeros(n)
dt = cfl*dx/c
print(dt,dx)
```

```python
# creating an array to store discrete values of x with step size dx
arrx = np.zeros(n)
print(arrx.shape)
dx1 = 0 
for i in range(n):
    arrx[i] = dx1
    dx1 += dx
arrx
```

```python
#initialising the value of u 
for i in range(n) :
        u[i] = math.sin(4 * math.pi* arrx[i]);
plt.plot(arrx,u)
```

```python
#for upwind approach
u_upwind = np.zeros(n)
flag = 0;
time = 0;
while(1):
    time = time + dt;
    u_old = u 
    if time>=1.5 and flag == 0:
        flag = 1
        u_upwind = u;
        plt.figure()
        plt.plot(arrx,u,color='r', label='t = 1.5')
        plt.xlabel("x")
        plt.ylabel("u")
        plt.title("Q2 upwind")
        plt.legend()
        plt.show()
    for j in range(1,n):
        u[j] = u_old[j] - cfl * (u_old[j] - u_old[j-1])
    #print(u)
    u[0] = u[n-2]
    #u[n-1] = u[2]
    if time > 2:
        break
        

```

```python
#for lax wendroff scheme FTCS
u_laxw = np.zeros(n)
flag = 0;
time = 0;
while(1):
    time = time + dt;
    u_old = u 
    #print(ti)
    if time>=1.5 and flag == 0:
        flag = 1
        u_laxw = u;
        plt.figure()
        plt.plot(arrx,u,color='r', label='t = 1.5')
        plt.xlabel("x")
        plt.ylabel("u")
        plt.title("Q2 Laxwendroff")
        plt.legend()
        plt.show()
        print("done")
    for j in range(1,n-1):
        u[j] = u_old[j] - cfl * 0.5*(u_old[j+1] - u_old[j-1]) + 0.5*cfl*cfl*(u_old[j+1] - 2*u_old[j] + u_old[j-1])
        
    u[n-1] = u_old[n-1] - cfl * (u_old[n-1] - u_old[n-2])
    u[0] = u_old[0]  - 0.5*cfl*(4*u_old[1] - u_old[2] - 3*u_old[i])
    #u[n-1] = u[2]
    if time > 2:
        break
        

```

```python
#for lax friedrichs scheme FTCS
u_laxf = np.zeros(n)
flag = 0;
time = 0;
while(1):
    time = time + dt;
    u_old = u 
    if time>=1.5 and flag == 0:
        flag = 1
        u_laxf = u;
        plt.figure()
        plt.plot(arrx,u,color='r', label='t = 1.5')
        plt.xlabel("x")
        plt.ylabel("u")
        plt.title("Lax fredrick")
        plt.legend()
        plt.show()
    for j in range(1,n-1):
        u[j] = (u_old[j+1] + u_old[j-1])/2 - cfl * (u_old[j+1] - u_old[j-1])/2
    u[n-1] = u_old[n-1] - cfl * (u_old[n-1] - u_old[n-2])
        
    #print(u)
    u[0] = u[n-1]
    #u[n-1] = u[2]
    if time > 2:
        break
        

```

```python
#analytical solution
u_anal = np.zeros(n)
for j in range(1,n-2):
        u_anal[j] = math.sin(4*math.pi*(arrx[j] - c*1.5))
plt.plot(arrx,u_anal)
```

```python
#plotting with matplotlib library on the same graph
plt.plot(arrx,u_anal,color='r', label='analytical')
plt.plot(arrx,u_upwind,color='y', label='upwind')
plt.plot(arrx,-1*u_laxw,color='g', label='lax wandroff')
plt.plot(arrx,u_laxf,color='b', label='lax fredireich')

plt.xlabel("x")
plt.ylabel("u")
plt.title("All graph comparison")
plt.legend()
plt.show()
```

# Q3 Part a
1. Note: run first three sections before running for any scheme 
2. We are taking the sig = 0.1 which gives us time step of 0.004

```python
#input paramenters
n = 101
x0 = -1
x1 = 1
dx =(x1-x0)/(n-1)

sig = 0.1
kvis =  1.5e-5
u = np.zeros(n)
u_old = np.zeros(n)
dt = sig*dx*2
print(dt,dx)
```

```python
# creating an array to store discrete values of x with step size dx
arrx = np.zeros(n)
print(arrx.shape)
dx1 = 0 
for i in range(n):
    arrx[i] = x0+dx1
    dx1 += dx
arrx
```

```python
#initialising the value of u 
for i in range(n) :
    if math.isclose(arrx[i], 0, abs_tol=1e-5):
        u[i] = 0.5
    elif arrx[i]>0:
        u[i] = 1
    elif arrx[i]<0:
        u[i] = 0
    

plt.plot(arrx,u)
```

```python
#for rachtmayer scheme, plotting at t = 0.04 
uhalf = np.zeros(n)
flag = 0;
time = 0;
while(1):
    time = time + dt;
    u_old = u 
    #print(ti)
    if time>=0.04 and flag == 0:
        flag = 1
        u_ract = u;
        plt.figure()
        plt.plot(arrx,u,color='r', label='t = 1.5')
        plt.xlabel("x")
        plt.ylabel("u")
        plt.title("Q3 ractmayer")
        plt.legend()
        plt.show()
        print("done")
    for j in range(1,n-1):
        uhalf[j] = (u_old[j+1] + u_old[j-1])/2 - 0.5*sig *u_old[j]* (u_old[j+1] - u_old[j-1]) + sig*(kvis/dx)*(u_old[j+1] - 2*u_old[j] + u_old[j-1])
        #print(uhalf)
    for j in range(1,n-1):    
        u[j] = u_old[j] - sig *uhalf[j]* (uhalf[j+1] - uhalf[j-1]) + 2*sig*(kvis/dx)*(uhalf[j+1] - 2*uhalf[j] + uhalf[j-1])
        
    u[n-1] = u_old[n-1] - 2*sig * (u[n-1] - u_old[n-2])
    u[0] = u[n-1]
    #u[n-1] = u[2]
    if time > 0.05:
        break
        

```

```python
#for macCormac scheme, plotting at t = 0.04 
ustar = np.zeros(n)
uhalf = np.zeros(n)
flag = 0;
time = 0;
while(1):
    time = time + dt;
    u_old = u 
    #print(ti)
    if time>=0.04 and flag == 0:
        flag = 1
        u_mac = u;
        plt.figure()
        plt.plot(arrx,u,color='r', label='t = 1.5')
        plt.xlabel("x")
        plt.ylabel("u")
        plt.title("Q3 mac Cormack")
        
        plt.show()
        print("done")
    for j in range(1,n-1):
        ustar[j] = u_old[j] - 2*sig *u_old[j]* (u_old[j+1] - u_old[j]) + 2*sig*(kvis/dx)*(u_old[j+1] - 2*u_old[j] + u_old[j-1])
        #print(uhalf)
    for j in range(1,n-1):
        uhalf[j] = (u_old[j] + ustar[j])/2
    for j in range(1,n-1):    
        u[j] = uhalf[j] - sig *ustar[j]* (ustar[j+1] - ustar[j-1]) + sig*(kvis/dx)*(ustar[j+1] - 2*ustar[j] + ustar[j-1])
        
    u[n-1] = u_old[n-1] - 2*sig * (u[n-1] - u_old[n-2])
    u[0] = u_old[0]  -sig*(4*u_old[1] - u_old[2] - 3*u_old[i])
    #u[n-1] = u[2]
    if time > 0.05:
        break
        

```

```python
#for beam warming scheme, plotting at t = 0.04 
flag = 0;
time = 0;
while(1):
    time = time + dt;
    u_old = u 
    #print(ti)
    if time>=0.04 and flag == 0:
        flag = 1
        u_beam = u;
        plt.figure()
        plt.plot(arrx,u,color='r', label='t = 1.5')
        plt.xlabel("x")
        plt.ylabel("u")
        plt.title("beam warming")
        
        plt.show()
        print("done")
    for j in range(1,n-1):
        u[j] = u_old[j] - sig*2*(3*u_old[j] - 4*u_old[j-1] + u_old[j-2]) + sig*sig*2*u_old[j]*u_old[j]*(u_old[j]-2*u_old[j-1] + u_old[j-2])+(kvis*2*sig/dx)*(u_old[j+1] - 2*u[j] + u[j-1])
    
        
    u[n-1] = u_old[n-1] - 2*sig * (u[n-1] - u_old[n-2])
    u[0] = u_old[0]  -sig*(4*u_old[1] - u_old[2] - 3*u_old[i])
    #u[n-1] = u[2]
    if time > 0.05:
        break
        

```
