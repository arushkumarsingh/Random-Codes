```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
import time
import pandas as pd
import numpy as np
import csv
```

```python
chromedriver_location = "D:\Setups\chromedriver"
driver = webdriver.Chrome(chromedriver_location)
driver.get('http://srivalab.cse.iitk.ac.in:3004/')
roll = '//*[@id="rollno"]'
submit = '/html/body/button'
```

```python
aa = np.arange(190001,191182)
aa
```

```python
with open('marks.csv','a',encoding  = 'utf-8',newline = '') as csv_file:
    writer = csv.writer(csv_file)
    for i in aa:
        try:
            driver.find_element_by_xpath(roll).clear()
            driver.find_element_by_xpath(roll).send_keys(str(i))
            driver.find_element_by_xpath(submit).click()
            WebDriverWait(driver,3).until(EC.alert_is_present(),'Timed out waiting for PA creation'+ 'conformation popup to appear.')
            driver.switch_to.alert.accept()

        except TimeoutException: 
            #driver.switchTo().alert().accept()
            #driver.find_element_by_xpath(roll).clear()
            #driver.find_element_by_xpath(roll).click()
            #driver.find_element_by_xpath(roll).send_keys(str(i))
            #driver.find_element_by_xpath(submit).click()
            name = driver.find_element_by_xpath('//*[@id="name"]').text
            section = driver.find_element_by_xpath('//*[@id="section"]').text
            minor = driver.find_element_by_xpath('//*[@id="mqsum"]').text
            lab = driver.find_element_by_xpath('//*[@id="lasum"]').text
            maj= driver.find_element_by_xpath('//*[@id="MQ"]').text
            midlab= driver.find_element_by_xpath('//*[@id="midsemlab"]').text
            midtheory = driver.find_element_by_xpath('//*[@id="midsem"]').text
            total = driver.find_element_by_xpath('//*[@id="total"]').text
            grade = driver.find_element_by_xpath('//*[@id="grade"]').text
            writer.writerow([name,section,minor,lab,maj,midlab,midtheory,total,grade])
```
